```python
import pandas as pd
from sklearn.metrics import confusion_matrix

# Calculate confusion matrices for the first step using Random Forest
cm_rf_cnn = confusion_matrix(y_test, prediction_1_rf)
cm_rf_vgg16 = confusion_matrix(y_test, prediction_1_rf_vgg16)
cm_rf_resnet = confusion_matrix(y_test, prediction_1_rf_resnet)
cm_rf_densenet = confusion_matrix(y_test, prediction_1_rf_densenet)


# Define a function to calculate metrics
def calculate_metrics(cm):
    TP = cm[1, 1]
    TN = cm[0, 0]
    FP = cm[0, 1]
    FN = cm[1, 0]
    total = TP + TN + FP + FN
    incorrect_fraction = (FP + FN) / total
    return TP, TN, FP, FN, incorrect_fraction


# Create a DataFrame to store the metrics for the first step using Random Forest
data_rf_first_step = {
    'Model': ['CNN_rf', 'VGG16_rf', 'ResNet_rf', 'DenseNet_rf'],
    'TP': [],
    'TN': [],
    'FP': [],
    'FN': [],
    'Fraction of Incorrect Predictions': [],
}

# Calculate metrics for the first step using Random Forest for each model
for cm, model in zip([cm_rf_cnn, cm_rf_vgg16, cm_rf_resnet, cm_rf_densenet], data_rf_first_step['Model']):
    TP, TN, FP, FN, incorrect_fraction = calculate_metrics(cm)
    data_rf_first_step['TP'].append(TP)
    data_rf_first_step['TN'].append(TN)
    data_rf_first_step['FP'].append(FP)
    data_rf_first_step['FN'].append(FN)
    data_rf_first_step['Fraction of Incorrect Predictions'].append(incorrect_fraction)

# Create a DataFrame for the first step using Random Forest
metrics_df_rf_first_step = pd.DataFrame(data_rf_first_step)

# Print the DataFrame for the first step using Random Forest
print("Random Forest Metrics - First Step:")
print(metrics_df_rf_first_step)
```
